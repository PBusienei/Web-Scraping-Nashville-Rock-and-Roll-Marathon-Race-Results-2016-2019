# Define the function 

def mar_table(pages, base, marathon_df):
   for page in range(1, pages+1):
       url = base + str(page)
       result = requests.post(url)
       soup = BS(result.text)
       tables = soup.find_all('table', attrs = {'class':'table table-responsive table-bordered'})
       results_list = pd.read_html(str(tables[0]))
       result_df = results_list[0]
       marathon_df = marathon_df.append(result_df)
   return marathon_df

```python
def mar_table(url, pgs):
    """Returns the dataframe
    pgs = number of pages 
    url = url
    """
    table = []
    for page, request in enumerate(range(pgs), start=1):
        result = requests.post(url+str(page))
        soup = BS(result.text)
        table.append(pd.read_html(str(soup))[1])
        
        return pd.concat(table)
    
```
